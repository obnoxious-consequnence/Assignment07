```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import bs4
import requests

driver = webdriver.Chrome()
driver.get("https://www.jobindex.dk")
driver.implicitly_wait(3)
search_field = driver.find_element_by_name('qs')
search_field.send_keys('Python')
search_field.submit()
sleep(3)
per = driver.find_element_by_class_name("close")
per.click()

r = requests.get(driver.current_url)
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')
events = soup.select('a b')
events2=[]

pages=8
for p in range(2,pages):
    nextPage=driver.find_element_by_link_text("{}".format(p))
    nextPage.click()        
    r = requests.get(driver.current_url)
    r.raise_for_status()
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    if p>=5:
        events2.extend(soup.select('a strong'))
    else:
        events.extend(soup.select('a b'))
    sleep(1)
jobs =events[::2]+events2    
print('Right now {} python jobs are available:'.format(len(jobs)))
for e in jobs:
    print(e.getText())
 
testSearch = driver.find_element_by_name('q')
testSearch.clear()
testSearch.send_keys('Kotlin')
testSearch.submit()
sleep(3)

r = requests.get(driver.current_url)
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')
kotlin = soup.select('a strong')

print('\nRight now {} Kotlin jobs are available:'.format(len(kotlin)))
for e in kotlin:
    print(e.getText())
```
